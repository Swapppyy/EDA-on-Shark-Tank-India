### Total number of Episodes

```python
%sql select max(epno) from data
```

### Unique brands in the show

```python
%sql select count(distinct brand) from data
```

### Pitches that got Funding

```python
%sql SELECT SUM(a.converted_not_converted) AS Pitches_funded, COUNT(*) \
FROM (SELECT amountinvestedlakhs, CASE WHEN amountinvestedlakhs > 0 THEN 1 ELSE 0 END AS converted_not_converted \
      FROM data) a;
```

### No. of male participants

```python
%sql select sum(male) from data
```

### No. of female participants

```python
%sql select sum(female) from data;
```

### Total amount invested by all the sharks in the show in the season 1

```python
%sql select sum(amountinvestedlakhs) from data;
```

### Average equity taken by the sharks

```python
%sql select avg(a.equitytakenp) from \
(select * from data where equitytakenp>0) a
```

### Highesh investment made by any of the shark

```python
%sql select max(amountinvestedlakhs) as max_amount from data;
```

### Highest Equity taken by any of the shark

```python
%sql select max(equitytakenp) from data
```

### Startups having at least one women

```python
%sql SELECT SUM(a.female_count) AS total_female_count \
FROM ( \
    SELECT female, COUNT(*) AS female_count \
    FROM data \
    WHERE female > 0 \
    GROUP BY female \
) AS a;
```

### Average teammembers

```python
%sql select avg(teammembers) as Average_teammembers from data
```

### Average amount invested per deal in the show

```python
%sql select avg(a.amountinvestedlakhs) amount_invested_per_deal from \
(select * from data where deal!='No Deal') a
```

### Average age group of contestants

```python
%sql select avgage,count(avgage) cnt from data group by avgage order by cnt desc
```

### Locations of the startups that pitched in the show

```python
%sql select location,count(location) cnt from data group by location order by cnt desc
```

### Different sectors from which the startups are

```python
%sql select sector,count(sector) cnt from data group by sector order by cnt desc
```

### Partner Deals of the sharks

```python
%sql select partners,count(partners) cnt from data  where partners!='NA' group by partners order by cnt desc
```

### Table of my 4 favorite sharks and their investments information

```python
import sqlite3

conn = sqlite3.connect('C:\\Users\\mahaj\\Downloads\\data.db')
cursor = conn.cursor()

query = """
SELECT m.keyy, m.total_deals_present, m.total_deals, n.total_amount_invested, n.avg_equity_taken
FROM (
    SELECT a.keyy, a.total_Deals_present, b.total_Deals
    FROM (
        SELECT 'Ashnner' AS keyy, COUNT(ashneeramountinvested) AS total_Deals_present
        FROM data
        WHERE ashneeramountinvested != 'NA'
    ) a
    INNER JOIN (
        SELECT 'Ashnner' AS keyy, COUNT(ashneeramountinvested) AS total_Deals
        FROM data
        WHERE ashneeramountinvested != 'NA' AND ashneeramountinvested != '0'
    ) b ON a.keyy = b.keyy
) m
INNER JOIN (
    SELECT 'Ashnner' AS keyy, SUM(c.ashneeramountinvested) AS total_amount_invested, AVG(c.ashneerequitytakenp) AS avg_equity_taken
    FROM (
        SELECT * FROM data WHERE ashneerequitytakenp != 'NA' AND ashneerequitytakenp != '0'
    ) c
) n ON m.keyy = n.keyy

UNION all

SELECT m.keyy, m.total_deals_present, m.total_deals, n.total_amount_invested, n.avg_equity_taken
FROM (
    SELECT a.keyy, a.total_Deals_present, b.total_Deals
    FROM (
        SELECT 'Peyush' AS keyy, COUNT(peyushamountinvested) AS total_Deals_present
        FROM data
        WHERE peyushamountinvested != 'NA'
    ) a
    INNER JOIN (
        SELECT 'Peyush' AS keyy, COUNT(peyushamountinvested) AS total_Deals
        FROM data
        WHERE peyushamountinvested != 'NA' AND peyushamountinvested != '0'
    ) b ON a.keyy = b.keyy
) m
INNER JOIN (
    SELECT 'Peyush' AS keyy, SUM(c.peyushamountinvested) AS total_amount_invested, AVG(c.peyushequitytakenp) AS avg_equity_taken
    FROM (
        SELECT * FROM data WHERE peyushequitytakenp != 'NA' AND peyushequitytakenp != '0'
    ) c
) n ON m.keyy = n.keyy

union all

SELECT m.keyy, m.total_deals_present, m.total_deals, n.total_amount_invested, n.avg_equity_taken
FROM (
    SELECT a.keyy, a.total_Deals_present, b.total_Deals
    FROM (
        SELECT 'Aman' AS keyy, COUNT(amanamountinvested) AS total_Deals_present
        FROM data
        WHERE amanamountinvested != 'NA'
    ) a
    INNER JOIN (
        SELECT 'Aman' AS keyy, COUNT(amanamountinvested) AS total_Deals
        FROM data
        WHERE amanamountinvested != 'NA' AND amanamountinvested != '0'
    ) b ON a.keyy = b.keyy
) m
INNER JOIN (
    SELECT 'Aman' AS keyy, SUM(c.amanamountinvested) AS total_amount_invested, AVG(c.amanequitytakenp) AS avg_equity_taken
    FROM (
        SELECT * FROM data WHERE amanequitytakenp != 'NA' AND amanequitytakenp != '0'
    ) c
) n ON m.keyy = n.keyy

union ALL

SELECT m.keyy, m.total_deals_present, m.total_deals, n.total_amount_invested, n.avg_equity_taken
FROM (
    SELECT a.keyy, a.total_Deals_present, b.total_Deals
    FROM (
        SELECT 'Anupam' AS keyy, COUNT(anupamamountinvested) AS total_Deals_present
        FROM data
        WHERE anupamamountinvested != 'NA'
    ) a
    INNER JOIN (
        SELECT 'Anupam' AS keyy, COUNT(anupamamountinvested) AS total_Deals
        FROM data
        WHERE anupamamountinvested != 'NA' AND anupamamountinvested != '0'
    ) b ON a.keyy = b.keyy
) m
INNER JOIN (
    SELECT 'Anupam' AS keyy, SUM(c.anupamamountinvested) AS total_amount_invested, AVG(c.anupamequitytakenp) AS avg_equity_taken
    FROM (
        SELECT * FROM data WHERE anupamequitytakenp != 'NA' AND anupamequitytakenp != '0'
    ) c
) n ON m.keyy = n.keyy;

"""

cursor.execute(query)

results = cursor.fetchall()
print(results)

conn.commit()
conn.close()
```

### Startups in which the highest amount has been invested in each domain/sector

```python
%sql SELECT c.*\
FROM (\
    SELECT brand,sector, amountinvestedlakhs, RANK() OVER (PARTITION BY sector ORDER BY amountinvestedlakhs DESC) AS rnk\
    FROM data\
) c\
WHERE c.rnk = 1;
```
